```python
from google_play_scraper import reviews_all, Sort
import pandas as pd
```

```python
APP_ID = 'com.miHoYo.GenshinImpact'
print("Mengambil data review...")

all_reviews = reviews_all(
    APP_ID,
    lang='id',
    country='id',
    sort=Sort.NEWEST
)

print("Total review berhasil diambil:", len(all_reviews))

dataset = []
for r in all_reviews:
    dataset.append({
        'review': r['content'],
        'rating': r['score'],
        'date': r['at']
    })
```

```python
df = pd.DataFrame(dataset)
df = df.dropna()

df.to_csv('genshin_full_review.csv', index=False, encoding='utf-8')
print("Data berhasil disimpan!")
df.head()
```
